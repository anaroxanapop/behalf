```python
import numpy as np, matplotlib.pyplot as plt, matplotlib as mpl, seaborn.apionly as sns
import plot_utils as pu
from glob import glob
%matplotlib inline
sns.set_context('poster')
mpl.rcParams['legend.fontsize'] = 20
mpl.rcParams['axes.labelsize'] = 20
# mpl.rcParams['ylabel.fontsize'] = 20
mpl.rcParams['xtick.labelsize'] = 20
mpl.rcParams['ytick.labelsize'] = 20
```

```python
sns.set_context('poster')
```

```python
results = {}
for infile in glob('../results/*'):
    run_name = infile.split('/')[-1]
    if (run_name.count('_') == 4) or ('serial' in run_name):
        try:
            results[run_name] = pu.RunResults(run_name)
            print(run_name)
        except FileNotFoundError:
            print('-- {:s}'.format(run_name))
        except BaseException as e:
            print('-------{:s}'.format(run_name))
            print(e)
runs = list(results.keys())
```

```python
serial_runs = [r for r in runs if 'serial' in r]
mpi_runs = [r for r in runs if 'mpi' in r]
gpu_runs = [r for r in runs if 'gpu_' in r]
gpuc_runs = [r for r in runs if 'gpuc' in r]
others = [r for r in runs if (r not in serial_runs) and (r not in mpi_runs) and (r not in gpu_runs) and (r not in gpuc_runs)]
```

```python
fig, axis = plt.subplots()
serial_times = []
serial_times.append(results['gpuc_1_1000_10_100'].med_times['Overall'])
serial_times.append(results['gpuc_1_4000_3_100'].med_times['Overall'])
serial_times.append(results['gpuc_1_16000_3_100'].med_times['Overall'])
Npartss = np.array([1000,4000,16000])
plt.plot(Npartss,serial_times,'ko-', label='1 core, 1 node, 1 GPU')
times_42 = []
times_42.append(results['gpuc_4_1000_10_100'].med_times['Overall'])
times_42.append(results['gpuc_4-2_4000_10_100'].med_times['Overall'])
times_42.append(results['gpuc_4-2_16000_3_100'].med_times['Overall'])
plt.plot(Npartss,times_42,'o-',label='4 cores, 2 nodes, 4 GPUs')
times_84 = []
times_84.append(results['gpuc_8-4_1000_10_100'].med_times['Overall'])
times_84.append(results['gpuc_8-4_4000_10_100'].med_times['Overall'])
times_84.append(results['gpuc_8-4_16000_3_100'].med_times['Overall'])
times_84.append(results['gpuc_8-4_100000_3_100'].med_times['Overall'])
Npartss = np.array([1000,4000,16000,100000])
plt.plot(Npartss,times_84,'o-',label='8 cores, 4 nodes, 8 GPUs')
times_168 = []
times_168.append(results['gpuc_16-8_1000_10_100'].med_times['Overall'])
times_168.append(results['gpuc_16-8_4000_10_100'].med_times['Overall'])
times_168.append(results['gpuc_16-8_16000_3_100'].med_times['Overall'])
times_168.append(results['gpuc_16-8_100000_3_100'].med_times['Overall'])
Npartss = np.array([1000,4000,16000,100000])
plt.plot(Npartss,times_168,'o-',label='16 cores, 8 nodes, 16 GPUs')
times_576 = []
times_576.append(results['gpuc_576-12_1000_10_100'].med_times['Overall'])
times_576.append(results['gpuc_576-12_4000_3_100'].med_times['Overall'])
times_576.append(results['gpuc_576-12_10000_3_100'].med_times['Overall'])
times_576.append(results['gpuc_576-12_16000_3_100'].med_times['Overall'])
times_576.append(results['gpuc_576-12_32000_3_100'].med_times['Overall'])
times_576.append(results['gpuc_576-12_100000_3_100'].med_times['Overall'])
Npartss = np.array([1000,4000,10000,16000,32000,100000])
plt.plot(Npartss,times_576,'co-',label='576 cores, 12 nodes, 24 GPUs')

'gpuc_576-12_10000_3_100'

yy = Npartss*np.log(Npartss)
yy2 = Npartss*Npartss
plt.plot(Npartss, serial_times[0]*yy/yy[0], 'mo--',label='N log(N)')
plt.plot(Npartss, serial_times[0]*yy2/yy2[0], 'ro:',label=r'$\mathrm{N}^2$')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.text(4e4,4e4,r'$\propto \mathrm{N}^2$',color='r',fontsize=20)
plt.text(4e4,1.5*1e3,r'$\propto \mathrm{N}$ log(N)',color='m',fontsize=20)
plt.ylabel('Total running time (s)')
plt.xlabel('N particles')
plt.tight_layout()
plt.savefig('TotalrunningTime_vs_Npart.png',dpi=300)
```

# Figure 5b

```python
fig, axis = plt.subplots()
times_1000 = []
baseline = results['gpuc_1_1000_10_100'].med_times['Overall']
times_1000.append(1.)
times_1000.append(baseline/4./results['gpuc_4_1000_10_100'].med_times['Overall'])
times_1000.append(baseline/8./results['gpuc_8-4_1000_10_100'].med_times['Overall'])
times_1000.append(baseline/16./results['gpuc_16-8_1000_10_100'].med_times['Overall'])
times_1000.append(baseline/576./results['gpuc_576-12_1000_10_100'].med_times['Overall'])
Ncores = np.array([1,4,8,16,576])
plt.plot(Ncores,times_1000,'-o',label='1000 particles')

times_1000 = []
baseline = results['gpuc_1_4000_3_100'].med_times['Overall']
times_1000.append(1.)
times_1000.append(baseline/4./results['gpuc_4-2_4000_10_100'].med_times['Overall'])
times_1000.append(baseline/8./results['gpuc_8-4_4000_10_100'].med_times['Overall'])
times_1000.append(baseline/16./results['gpuc_16-8_4000_10_100'].med_times['Overall'])
times_1000.append(baseline/576./results['gpuc_576-12_4000_3_100'].med_times['Overall'])
Ncores = np.array([1,4,8,16,576])
plt.plot(Ncores,times_1000,'-o',label='4000 particles')

times_1000 = []
baseline = results['gpuc_1_16000_3_100'].med_times['Overall']
times_1000.append(1.)
times_1000.append(baseline/4./results['gpuc_4-2_16000_3_100'].med_times['Overall'])
times_1000.append(baseline/8./results['gpuc_8-4_16000_3_100'].med_times['Overall'])
times_1000.append(baseline/16./results['gpuc_16-8_16000_3_100'].med_times['Overall'])
times_1000.append(baseline/576./results['gpuc_576-12_16000_3_100'].med_times['Overall'])
Ncores = np.array([1,4,8,16,576])
plt.plot(Ncores,times_1000,'-o',label='16000 particles')
plt.plot(Ncores,1./Ncores,'--')
plt.xscale('log')
plt.yscale('log')
plt.legend()
props = dict(boxstyle='round', facecolor='w', alpha=0.5)
plt.text(1e0,6e-3,'These were runs done with:\n 4    cores on   2 nodes  + 4   GPUs \n 8    cores on   4 nodes  + 8   GPUs \n 16  cores on   8 nodes  + 16 GPUs \n 576 cores on 12 nodes + 24 GPUs',color='k',fontsize=16,bbox=props)
plt.text(1e2,2e-3,r'$\propto 1/\mathrm{N}_{\mathrm{cores}}$',color='r',fontsize=20)
plt.ylabel('Efficiency')
plt.xlabel('N cores')
plt.tight_layout
plt.savefig('Efficiency_vs_Npart.png',dpi=300)
```
